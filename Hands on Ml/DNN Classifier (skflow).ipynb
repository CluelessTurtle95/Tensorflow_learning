```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
data = pd.read_csv("mnist_train.csv")
```

```python
data.head()
```

```python
ym = data["label"]
Xm = data.drop(columns="label" , inplace=False)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

Xs = scaler.fit_transform(Xm.values)
ys = ym.values
```

```python
feature_col = tf.contrib.learn.infer_real_valued_columns_from_input(Xs)
clf = tf.contrib.learn.DNNClassifier(hidden_units=[300,100] , feature_columns=feature_col , n_classes=10)
```

```python
dnn_clf = tf.contrib.learn.SKCompat(clf)
```

```python
dnn_clf.fit(Xs , ys , batch_size = 100 , steps= 10000)
```

```python
test = pd.read_csv("mnist_test.csv")
ytest = test["label"].values
Xtest = test.drop(columns="label" , inplace=False).values
Xtest = scaler.transform(Xtest)
```

```python
from sklearn.metrics import accuracy_score

ypred = dnn_clf.predict(Xtest)

accuracy_score(ytest , ypred['classes'])
```
